
#### [4.Tibetan](https://github.com/yongtso/TSTD)



```python
! wget https://raw.githubusercontent.com/yongtso/TSTD/master/TSTD.txt
file = open("TSTD.txt","r",encoding = "utf-8")
lines = file.readlines()
texts= []
y = []
for line in lines : 
	if "+1" in line:
		y.append("positive")
		texts.append(line.replace("+1",""))
	elif "-1" in line:
		y.append("negative")
		texts.append(line.replace("-1",""))
train_df = {"text" :[] ,"y" : []}
train_df= pd.DataFrame(data=train_df)
train_df["text"]=texts
train_df["y"] = y
train_df = train_df[train_df["y"].str.contains("positive")  | train_df["y"].str.contains("negative")]
tibetan_df = train_df
tibetan_df
```
